```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
X = df.drop(columns='diagnosis')
y = df['diagnosis']

```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
```

```python
log1 = LogisticRegression()

cv_scores = cross_val_score(log1,X,y,cv=5)
cv_scores
```

```python
print(cv_scores.mean())
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.2,random_state=0)
```

```python
print(X_train.shape,X_test.shape)
```

```python
scaler = MinMaxScaler(clip = True)
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Print the min and max values of the scaled data
print("Training set min:", X_train_scaled.min())
print("Training set max:", X_train_scaled.max())
print("Test set min:", X_test_scaled.min())
print("Test set max:", X_test_scaled.max())
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
kbest = SelectKBest(chi2, k=25)
X_train_kbest = kbest.fit_transform(X_train_scaled, y_train)
X_test_kbest = kbest.fit_transform(X_test_scaled, y_test)
selected_features = kbest.get_support(indices=True)

selected_features
```

```python
log = LogisticRegression()
log
```

```python
log.fit(X_train_kbest,y_train)
```

```python
print("Log Accuracy on train data",log.score(X_train_kbest,y_train))
print("Log Accuracy on test data ",log.score(X_test_kbest,y_test))
```
